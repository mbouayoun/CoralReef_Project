```python
# import the required library
import numpy as np
import cv2
from matplotlib import pyplot as plt
  
def findCorners(image_file):  
# read the image
    img = cv2.imread(image_file)  
    # convert image to gray scale image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  
    # detect corners with the goodFeaturesToTrack function.
    corners = cv2.goodFeaturesToTrack(gray, 27, 0.01, 10)
    corners = np.int0(corners)

    rect = cv2.minAreaRect(corners) # get a rectangle rotated to have minimal area
    box = cv2.boxPoints(rect) # get the box from the rectangle
    box = np.int0(box) # the box is now the new contour.
    p1 = [box[0][0],box[0][1]]
    p2 = [box[1][0],box[1][1]]
    p3 = [box[2][0],box[2][1]]
    p4 = [box[3][0],box[3][1]]

    return (img, p1, p2, p3, p4)

image,a,b,c,d = findCorners('./data/brute/Test/E2B_07_2020_rot.jpg')
```

```python
# Green color in BGR
color = (0, 255, 0)
  
# Line thickness of 9 px
thickness = 9

image=cv2.circle(image,a, 30,color, thickness)
image=cv2.circle(image,b, 30,color, thickness)
image=cv2.circle(image,c, 30,color, thickness)
image=cv2.circle(image,d, 30,color, thickness)
```

```python
plt.imshow(image)
plt.show()
```
